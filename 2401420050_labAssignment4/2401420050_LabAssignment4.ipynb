## N-Queens(Backtracking)

```python
def is_safe(board, row, col, n):
    for i in range(col):
        if board[row][i] == 1:
            return False

    i, j = row, col
    while i >= 0 and j >= 0:
        if board[i][j] == 1:
            return False
        i -= 1
        j -= 1

    i, j = row, col
    while i < n and j >= 0:
        if board[i][j] == 1:
            return False
        i += 1
        j -= 1

    return True


def solve_nq_util(board, col, n):
    if col >= n:
        return True

    for i in range(n):
        if is_safe(board, i, col, n):
            board[i][col] = 1

            if solve_nq_util(board, col + 1, n):
                return True

            board[i][col] = 0

    return False


def solve_n_queens(n):
    board = [[0]*n for _ in range(n)]

    if not solve_nq_util(board, 0, n):
        print("No solution")
        return

    for row in board:
        print(row)


solve_n_queens(4)
```

## Subset_Problem

```python
def subset_sum(nums, target, path=[], index=0):
    if target == 0:
        print(path)
        return

    if target < 0 or index >= len(nums):
        return

    # Include
    subset_sum(nums, target - nums[index], path + [nums[index]], index + 1)

    # Exclude
    subset_sum(nums, target, path, index + 1)


nums = [2, 3, 5, 7]
target = 7

subset_sum(nums, target)
```

## Naive String Matching 

```python
def naive_string_match(text, pattern):
    n = len(text)
    m = len(pattern)

    for i in range(n - m + 1):
        if text[i:i+m] == pattern:
            print("Pattern found at index", i)


text = "ABABDABACDABABCABAB"
pattern = "ABABCABAB"

naive_string_match(text, pattern)
```

## Finite Automata (Basic Simulation)

```python
def finite_automata_search(text, pattern):
    m = len(pattern)
    n = len(text)

    for i in range(n - m + 1):
        state = 0
        for j in range(m):
            if text[i + j] == pattern[j]:
                state += 1
            else:
                break

        if state == m:
            print("Pattern found at index", i)


finite_automata_search("AABAACAADAABAABA", "AABA")
```
