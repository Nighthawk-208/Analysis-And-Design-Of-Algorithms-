## BFS

```python
from collections import deque

def bfs(graph, start):
    visited = set()
    queue = deque([start])
    visited.add(start)
    order = []

    while queue:
        node = queue.popleft()
        order.append(node)

        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)

    return order


graph = {
    'Alice': ['Bob', 'Charlie'],
    'Bob': ['Alice', 'David', 'Eve'],
    'Charlie': ['Alice', 'David'],
    'David': ['Bob', 'Charlie', 'Eve'],
    'Eve': ['Bob', 'David']
}

print("BFS:", bfs(graph, 'Alice'))
```

## DFS

```python
def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()

    visited.add(start)
    order = [start]

    for neighbor in graph[start]:
        if neighbor not in visited:
            order += dfs(graph, neighbor, visited)

    return order


print("DFS:", dfs(graph, 'Alice'))
```

## Topological Sort 

```python
from collections import deque

def topological_sort(graph):
    in_degree = {u: 0 for u in graph}

    for u in graph:
        for v in graph[u]:
            in_degree[v] = in_degree.get(v, 0) + 1

    queue = deque([u for u in in_degree if in_degree[u] == 0])
    topo_order = []

    while queue:
        node = queue.popleft()
        topo_order.append(node)

        for neighbor in graph[node]:
            in_degree[neighbor] -= 1
            if in_degree[neighbor] == 0:
                queue.append(neighbor)

    return topo_order


task_graph = {
    'A': ['C'],
    'B': ['C'],
    'C': ['D'],
    'D': ['E'],
    'E': []
}

print("Topological Order:", topological_sort(task_graph))
```

## Dijkstra

```python
import heapq

def dijkstra(graph, start):
    dist = {node: float('inf') for node in graph}
    dist[start] = 0

    pq = [(0, start)]
    prev = {node: None for node in graph}

    while pq:
        d, u = heapq.heappop(pq)

        if d > dist[u]:
            continue

        for v, w in graph[u]:
            if dist[u] + w < dist[v]:
                dist[v] = dist[u] + w
                prev[v] = u
                heapq.heappush(pq, (dist[v], v))

    return dist, prev


def reconstruct_path(prev, end):
    path = []
    while end:
        path.append(end)
        end = prev[end]
    return path[::-1]


city_graph = {
    'Hospital': [('A', 4), ('B', 2)],
    'A': [('Hospital', 4), ('B', 1), ('C', 5)],
    'B': [('Hospital', 2), ('A', 1), ('D', 3)],
    'C': [('A', 5), ('D', 2), ('Accident', 1)],
    'D': [('B', 3), ('C', 2), ('Accident', 4)],
    'Accident': [('C', 1), ('D', 4)]
}

dist, prev = dijkstra(city_graph, 'Hospital')
print("Distances:", dist)
print("Path:", reconstruct_path(prev, 'Accident'))
```

## Bellman Ford

```python
def bellman_ford(graph, vertices, start):
    dist = {v: float('inf') for v in vertices}
    dist[start] = 0

    edges = [(u, v, w) for u in graph for v, w in graph[u]]

    for _ in range(len(vertices) - 1):
        for u, v, w in edges:
            if dist[u] + w < dist[v]:
                dist[v] = dist[u] + w

    for u, v, w in edges:
        if dist[u] + w < dist[v]:
            print("Negative cycle detected")
            return None

    return dist


route_graph = {
    'S': [('A', 4), ('B', 2)],
    'A': [('C', -1)],
    'B': [('A', 1), ('C', 3)],
    'C': [('D', 2)],
    'D': []
}

vertices = ['S', 'A', 'B', 'C', 'D']

print(bellman_ford(route_graph, vertices, 'S'))
```

## Prim’s Algorithm

```python
import heapq

def prim(graph, start):
    visited = set([start])
    edges = [(w, start, v) for v, w in graph[start]]
    heapq.heapify(edges)

    mst = []
    total_cost = 0

    while edges:
        w, u, v = heapq.heappop(edges)

        if v not in visited:
            visited.add(v)
            mst.append((u, v, w))
            total_cost += w

            for neighbor, weight in graph[v]:
                if neighbor not in visited:
                    heapq.heappush(edges, (weight, v, neighbor))

    return mst, total_cost


cable_graph = {
    'Router': [('Floor1', 4), ('Floor2', 2)],
    'Floor1': [('Router', 4), ('Floor2', 1), ('Floor3', 5)],
    'Floor2': [('Router', 2), ('Floor1', 1), ('Floor4', 3)],
    'Floor3': [('Floor1', 5), ('Floor4', 2), ('Server', 1)],
    'Floor4': [('Floor2', 3), ('Floor3', 2), ('Server', 4)],
    'Server': [('Floor3', 1), ('Floor4', 4)]
}

print(prim(cable_graph, 'Router'))
```

## Kruskal

```python
def find(parent, x):
    if parent[x] != x:
        parent[x] = find(parent, parent[x])
    return parent[x]


def union(parent, rank, x, y):
    rx, ry = find(parent, x), find(parent, y)

    if rx == ry:
        return False

    if rank[rx] < rank[ry]:
        rx, ry = ry, rx

    parent[ry] = rx

    if rank[rx] == rank[ry]:
        rank[rx] += 1

    return True


def kruskal(vertices, edges):
    edges.sort(key=lambda e: e[2])

    parent = {v: v for v in vertices}
    rank = {v: 0 for v in vertices}

    mst = []
    total = 0

    for u, v, w in edges:
        if union(parent, rank, u, v):
            mst.append((u, v, w))
            total += w

    return mst, total


vertices_k = ['Router','Floor1','Floor2','Floor3','Floor4','Server']

edges_k = [
    ('Router','Floor1',4), ('Router','Floor2',2),
    ('Floor1','Floor2',1), ('Floor1','Floor3',5),
    ('Floor2','Floor4',3), ('Floor3','Floor4',2),
    ('Floor3','Server',1), ('Floor4','Server',4)
]

print(kruskal(vertices_k, edges_k))
```
